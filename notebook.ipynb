```python
chi_crime = pd.read_csv("Crimes_-_2019.csv")
chi_crime.head()
```

```python
crime_dates = []
for x in chi_crime['Date']:
    crime_dates.append(x[6:10]+"-"+x[0:2]+"-"+x[3:5])
crime_dates = pd.DataFrame(crime_dates)
```

```python
chi_crime.insert(len(chi_crime.columns),'DATE', crime_dates)
```

```python
chi_crime.head()
```

```python
total_crimes = []
total_arrests = []
for d in chi_weath['DATE']:
    temp = chi_crime.loc[chi_crime['DATE'] == d]
    total_crimes.append(len(temp))
    total_arrests.append(len(temp.loc[temp['Arrest']==True]))
```

```python
chi_crime_weather = pd.DataFrame()
chi_crime_weather.insert(0, 'DATE', chi_weath['DATE'])
chi_crime_weather.insert(1, 'TEMP', chi_weath['TAVG'])
chi_crime_weather.insert(2, 'PRCP', chi_weath['PRCP'])
chi_crime_weather.insert(3, 'SNOW', chi_weath['SNOW'])
chi_crime_weather.insert(4, 'WIND', chi_weath['AWND'])
chi_crime_weather.insert(5, 'CRIMES', total_crimes)
chi_crime_weather.insert(6, 'ARRESTS', total_arrests)

chi_crime_weather = chi_crime_weather.reset_index(drop=True)
chi_crime_weather
```
