```python
import pandas as pd
import numpy as np
```

```python
chi_weather19 = pd.read_csv('chicago_weather_19.csv') ## contains the weather data from 2019
chi_weather18 = pd.read_csv('chicago_weather_18.csv') ## contains the weather data from 2018
chi_weather17 = pd.read_csv('chicago_weather_16_17.csv') ## contains the weather data from 2016-2017

chi_crime19 = pd.read_csv('Crimes_-_2019.csv')
chi_crime18 = pd.read_csv('Crimes_-_2018.csv')
chi_crime17 = pd.read_csv('Crimes_-_2017.csv')
chi_crime16 = pd.read_csv('Crimes_-_2016.csv')
```

```python
## making sample smaller by only using one station: USW00094846 aka O'Hare International Airport

chi_weath = chi_weather19[39182:] ## 2019
chi_weath2 = chi_weather18[39733:] ## 2018
chi_weath3 = chi_weather17[76435:] ## 2016-2017
```

```python
crime_dates_19 = []
for x in chi_crime19['Date']:
    crime_dates_19.append(x[6:10]+"-"+x[0:2]+"-"+x[3:5])
crime_dates_19 = pd.DataFrame(crime_dates_19)

crime_dates_18 = []
for x in chi_crime18['Date']:
    crime_dates_18.append(x[6:10]+"-"+x[0:2]+"-"+x[3:5])
crime_dates_18 = pd.DataFrame(crime_dates_18)

crime_dates_17 = []
for x in chi_crime17['Date']:
    crime_dates_17.append(x[6:10]+"-"+x[0:2]+"-"+x[3:5])
crime_dates_17 = pd.DataFrame(crime_dates_17)

crime_dates_16 = []
for x in chi_crime16['Date']:
    crime_dates_16.append(x[6:10]+"-"+x[0:2]+"-"+x[3:5])
crime_dates_16 = pd.DataFrame(crime_dates_16)
```

```python
chi_crime19.insert(len(chi_crime19.columns),'DATE', crime_dates_19)
chi_crime18.insert(len(chi_crime18.columns),'DATE', crime_dates_18)
chi_crime17.insert(len(chi_crime17.columns),'DATE', crime_dates_17)
chi_crime16.insert(len(chi_crime16.columns),'DATE', crime_dates_16)
```

```python
chi_crime19.head()
chi_crime18.head()
chi_crime17.head()
chi_crime16.head()
```

```python
total_crimes_19 = []
total_arrests_19 = []
total_crimes_18 = []
total_arrests_18 = []
total_crimes_1617 = []
total_arrests_1617 = []

for d in chi_weath['DATE']:
    temp19 = chi_crime19.loc[chi_crime19['DATE'] == d]
    total_crimes_19.append(len(temp19))
    total_arrests_19.append(len(temp19.loc[temp19['Arrest']==True]))
    
for e in chi_weath2['DATE']:
    temp18 = chi_crime18.loc[chi_crime18['DATE'] == e]
    total_crimes_18.append(len(temp18))
    total_arrests_18.append(len(temp18.loc[temp18['Arrest']==True]))
    
for f in chi_weath3['DATE']:
    if f[0:4] == "2016":
        temp16 = chi_crime16.loc[chi_crime16['DATE'] == f]
        total_crimes_1617.append(len(temp16))
        total_arrests_1617.append(len(temp16.loc[temp16['Arrest']==True]))
    else:
        temp17 = chi_crime17.loc[chi_crime17['DATE'] == f]
        total_crimes_1617.append(len(temp17))
        total_arrests_1617.append(len(temp17.loc[temp17['Arrest']==True]))
```

```python
chi_crime_weather = pd.DataFrame()
chi_crime_weather.insert(0, 'DATE', chi_weath['DATE'])
chi_crime_weather.insert(1, 'TEMP', chi_weath['TAVG'])
chi_crime_weather.insert(2, 'PRCP', chi_weath['PRCP'])
chi_crime_weather.insert(3, 'SNOW', chi_weath['SNOW'])
chi_crime_weather.insert(4, 'WIND', chi_weath['AWND'])
chi_crime_weather.insert(5, 'CRIMES', total_crimes_19)
chi_crime_weather.insert(6, 'ARRESTS', total_arrests_19)

chi_crime_weather = chi_crime_weather.reset_index(drop=True)
chi_crime_weather
```
